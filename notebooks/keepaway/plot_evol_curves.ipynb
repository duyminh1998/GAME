```python
from GAME.utils.config import config
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

config_data = config()
```

```python
keepaway_GAME_evol_folder_path = os.path.join(config_data['output_path'], '11122022 Evolve keepaway Maps with GAME 2000 FE')
trials = 10
GAME_fitness_evaluations = [i for i in range(20, 2020, 20)]
GAME_trials_best_fitness = []
for trial in range(trials):
    trial_stats = os.path.join(keepaway_GAME_evol_folder_path, 'trial{}_stats.txt'.format(trial))
    trial_df = pd.read_csv(trial_stats, index_col=False, header=0)    
    if len(np.array(trial_df['best_fitness'].to_list())) == 99:
        GAME_trials_best_fitness.append(np.array(trial_df['best_fitness'].to_list()))
GAME_trials_best_fitness = np.array(GAME_trials_best_fitness, dtype=object)
GAME_trials_best_fitness = GAME_trials_best_fitness.mean(axis = 0)
GAME_trials_best_fitness = [0.35] + list(GAME_trials_best_fitness)

keepaway_GAMERMHC_evol_folder_path = os.path.join(config_data['output_path'], '11122022 Evolve keepaway Maps with GAME_RMHC 2000 FE')
trials = 10
RMHC_fitness_evaluations = np.arange(2, 2001, 1, dtype=int)
RMHC_trials_best_fitness = []
for trial in range(trials):
    trial_stats = os.path.join(keepaway_GAMERMHC_evol_folder_path, 'trial{}_stats.txt'.format(trial))
    trial_df = pd.read_csv(trial_stats, index_col=False, header=0)    
    RMHC_trials_best_fitness.append(trial_df['best_fitness'].to_list())
RMHC_trials_best_fitness = np.array(RMHC_trials_best_fitness)
RMHC_trials_best_fitness = RMHC_trials_best_fitness.mean(axis = 0)
```

```python
plt.plot(GAME_fitness_evaluations, GAME_trials_best_fitness, label='GAME')
plt.plot(RMHC_fitness_evaluations, RMHC_trials_best_fitness, label='GAME-RMHC', alpha=0.5)
plt.legend()
plt.ylim(0, 1.1)
plt.xlim(0, 2000)
plt.xticks([i for i in range(0, 2000, 200)])
plt.yticks([i*0.1 for i in range(0, 11, 1)])
plt.xlabel('Fitness evaluations')
plt.ylabel('Best fitness')
plt.title('Evolutionary curves for GAME and GAME-RMHC on Keepaway')
plt.show()
```
